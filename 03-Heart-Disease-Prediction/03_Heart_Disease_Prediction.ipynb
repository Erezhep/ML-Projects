# 📊 Логистикалық Регрессия: Толық Ақпараттық Шпаргалка

**Логистикалық регрессия (Logistic Regression)** — бұл жіктеу (classification) есептерін шешуге арналған ең танымал әрі іргелі машиналық оқыту алгоритмдерінің бірі. Ол объектінің белгілі бір класқа жату **ықтималдығын** (probability) болжайды.

---

### 1. Негізгі Концепция
* **Негізгі мақсаты:** Классификация (мысалы: Хат спам ба, әлде жоқ па?).
* **Шығыс мәні (Output):** 0 мен 1 аралығындағы сан (ықтималдық).
* **Шешім шекарасы (Decision Boundary):** Әдетте $0.5$ (яғни $50\%$).
  * Егер $P \ge 0.5 \rightarrow$ Класс 1.
  * Егер $P < 0.5 \rightarrow$ Класс 0.

---

### 2. Математикалық Моделі

Логистикалық регрессия екі қадамнан тұрады: **Сызықтық теңдеу** және оны **Сигмоид функциясына** салу.

**Қадам 1. Сызықтық комбинация (Log-odds):**
Модель алдымен барлық белгілерді ($x$) өз салмақтарына ($w$) көбейтіп, еркін мүшені ($b$) қосады:
$$z = w_1x_1 + w_2x_2 + \dots + w_nx_n + b$$

**Қадам 2. Sigmoid (Логистикалық) функциясы:**
Алынған $z$ мәнін $[0, 1]$ интервалына сыйдыру үшін Сигмоид функциясы қолданылады:
$$\sigma(z) = \frac{1}{1 + e^{-z}}$$
*(Мұндағы $e \approx 2.718$ — Эйлер саны).*



---

### 3. Шығын Функциясы (Cost Function)

Сызықтық регрессиядағыдай (MSE) қателікті есептеу мұнда жарамайды. Логистикалық регрессия үшін **Log Loss** (немесе Binary Cross-Entropy) қолданылады:

$$J(w,b) = - \frac{1}{m} \sum_{i=1}^{m} \left[ y^{(i)} \log(\hat{y}^{(i)}) + (1 - y^{(i)}) \log(1 - \hat{y}^{(i)}) \right]$$

* $m$: Деректер саны.
* $y$: Нақты мән (0 немесе 1).
* $\hat{y}$: Модельдің болжаған ықтималдығы.
* **Мақсат:** Осы $J(w,b)$ мәнін минимумға жеткізу (Градиентті түсу - Gradient Descent арқылы).

---

### 4. Логистикалық Регрессияның Түрлері

| Түрі | Сипаттамасы | Мысал |
| :--- | :--- | :--- |
| **Бинарлы (Binary)** | Тек 2 класс бар. | Ауру / Сау, Спам / Спам емес. |
| **Көпкласты (Multinomial)** | 3 немесе одан көп класс (реті маңызды емес). | Ит / Мысық / Құс. |
| **Реттік (Ordinal)** | 3 немесе одан көп класс (реті маңызды). | Нашар / Орташа / Жақсы. |

---

### 5. Алгоритмнің Артықшылықтары мен Кемшіліктері

**Артықшылықтары (Pros):**
* **Қарапайымдылық:** Түсінуге және түсіндіруге өте оңай.
* **Жылдамдық:** Есептеу ресурстарын аз талап етеді.
* **Интерпретация:** Салмақ коэффициенттері ($w$) қай белгінің (feature) қаншалықты маңызды екенін көрсетеді.
* **Ықтималдық:** Тек "иә/жоқ" емес, нақты пайызбен болжам береді.

**Кемшіліктері (Cons):**
* **Сызықтық шектеу:** Деректер сызықтық түрде бөлінбейтін болса (Non-linear), нашар жұмыс істейді.
* **Мультиколлинеарлыққа сезімтал:** Тәуелсіз айнымалылар бір-бірімен қатты байланысты болмауы керек.
* Күрделі байланыстарды (мысалы, сурет немесе мәтін тану) ұстай алмайды.

---

### 6. Қолданылатын Негізгі Метрикалар
Модельдің сапасын бағалау үшін мына метрикалар қолданылады:
1. **Confusion Matrix (Шатасу матрицасы):** Дұрыс және қате болжамдардың 4 түрін (TP, TN, FP, FN) көрсетеді.
2. **Accuracy:** Жалпы дұрыс табылған жауаптардың үлесі.
3. **Precision & Recall:** Баланссыз деректер (Imbalanced data) үшін маңызды.
4. **ROC-AUC қисығы:** Модельдің екі класты бір-бірінен ажырату қабілеті (1-ге жақын болған сайын жақсы).

# 📈 Классификация Метрикалары

Классификация есептерінде модельдің қаншалықты жақсы жұмыс істейтінін бағалау үшін бірнеше метрикалар қолданылады. Олардың барлығы дерлік **Confusion Matrix**-тен бастау алады.

---

### 1. Confusion Matrix (Шатасу матрицасы)
Бұл — модельдің дұрыс және қате болжамдарын көрсететін 2x2 өлшемді кесте (бинарлы классификация үшін).

image.png
---

| | Болжам: Оң (1) | Болжам: Теріс (0) |
| :--- | :--- | :--- |
| **Нақты: Оң (1)** | **TP** (True Positive) | **FN** (False Negative) |
| **Нақты: Теріс (0)** | **FP** (False Positive) | **TN** (True Negative) |

* **TP (True Positive - Шын оң):** Модель «Ауру» деді, адам шынымен де ауру. (Керемет!)
* **TN (True Negative - Шын теріс):** Модель «Сау» деді, адам шынымен де сау. (Керемет!)
* **FP (False Positive - Жалған оң / 1-ші текті қате):** Модель «Ауру» деді, бірақ адам сау. (Өтірік дабыл).
* **FN (False Negative - Жалған теріс / 2-ші текті қате):** Модель «Сау» деді, бірақ адам ауру. (Ең қауіпті қате, себебі ауруды өткізіп алдық).

---

### 2. Accuracy (Жалпы дәлдік)
Барлық жағдайлардың ішінен модель қаншасын дұрыс тапқанын көрсетеді.

**Формула:**
$$Accuracy = \frac{TP + TN}{TP + TN + FP + FN}$$

* **Қашан қолдануға болады?** Кластар саны бірдей (балансты) болған кезде.
* **Қашан қолдануға БОЛМАЙДЫ?** Деректер баланссыз (Imbalanced) болғанда. Мысалы, 100 адамның 99-ы сау, 1-еуі ғана ауру болса, модель бәрін "Сау" деп тапса да, Accuracy 99% болады, бірақ ол жалғыз ауруды таба алмағандықтан, модель түкке тұрғысыз.

---

### 3. Precision және Recall (Нақтылық пен Толықтық)

Бұл екеуі әсіресе баланссыз деректермен жұмыс істегенде маңызды.

image.png

#### A) Precision (Нақтылық)
Модель "Оң (1)" деп тапқандардың қаншасы шынымен де "Оң"? Яғни, модель өз сөзіне қаншалықты жауап береді?

**Формула:**
$$Precision = \frac{TP}{TP + FP}$$

* **Мақсаты:** False Positive (өтірік дабыл) санын азайту.
* **Мысал:** Спам-фильтр. Маңызды хатты спам деп (FP) өшіріп тастамау өте маңызды. Сондықтан мұнда Precision жоғары болуы керек.

#### B) Recall / Sensitivity (Толықтық)
Нақты "Оң (1)" кластардың қаншасын біздің модель тауып алды?

**Формула:**
$$Recall = \frac{TP}{TP + FN}$$

* **Мақсаты:** False Negative (өткізіп алу) санын азайту.
* **Мысал:** Қатерлі ісікті (рак) анықтау. Сау адамды рак деп қателесу (FP) қорқынышты емес (қосымша анализ тапсырады), ал ракпен ауыратын адамды "сау" деп (FN) үйіне жіберу өте қауіпті. Мұнда Recall жоғары болуы шарт.

---

### 4. F1-Score (F1-өлшемі)
Precision мен Recall арасындағы компромисс (гармониялық орташа мән). Екеуі де жоғары болғанда ғана F1-Score жоғары болады.

**Формула:**
$$F1 = 2 \times \frac{Precision \times Recall}{Precision + Recall}$$

* **Қашан қолданамыз?** Деректер баланссыз болғанда және бізге Precision мен Recall-дың ортақ бір бағасы керек болғанда.

---

### 5. ROC Curve және AUC (ROC-AUC қисығы)
Бұл — модельдің екі класты бір-бірінен ажырату (дискриминациялау) қабілетін көрсететін график және оның ауданы.

image.png

* **ROC қисығы:** Барлық мүмкін болатын шектік мәндер (thresholds) үшін **TPR** (True Positive Rate = Recall) мен **FPR** (False Positive Rate) тәуелділігін көрсетеді.
* **AUC (Area Under the Curve):** Осы ROC қисығының астындағы аудан.

**AUC мәнін қалай түсінеміз?**
* **$AUC = 1.0$:** Идеалды модель (қатесіз табады).
* **$AUC \ge 0.8$:** Өте жақсы модель.
* **$AUC \approx 0.5$:** Модельдің ешқандай болжау қабілеті жоқ (тиын лақтырып тапқанмен бірдей).
* **$AUC < 0.5$:** Модель керісінше жұмыс істеп тұр (ауруды сау, сауды ауру дейді).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                             f1_score, confusion_matrix, roc_curve, auc)
```

```python
path = "/content/heart.csv"

# Деректерді жүктеу
data = pd.read_csv(path)  # 0 - сау, 1 - жүрек ауруы бар
data.head()
```

```python
# Деректер өлшемі
data.shape
```

```python
# Деректер туралы ақпарат
data.info()
```

```python
# Деректерге қысқаша сипаттама
data.describe()
```

```python
# Корреляция
sns.heatmap(data.corr())
plt.show()
```

```python
# Деректерді бөлу
X = data.drop('target', axis=1)
y = data['target']

# НОРМАЛИЗАЦИЯ (Масштабтау)
'''
Логистикалық регрессия үшін барлық сандарды (мысалы,
200 холестерин мен 1.5 ST депрессиясы) бір масштабқа әкелу өте маңызды
 '''
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# TRAIN / TEST БӨЛУ (80% оқытуға, 20% тексеруге)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=20)
```

```python
# МОДЕЛЬДІ ҚҰРУ ЖӘНЕ ОҚЫТУ
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# БОЛЖАМДАР ЖАСАУ
y_pred = model.predict(X_test)             # Кластар: 0 немесе 1
y_prob = model.predict_proba(X_test)[:, 1] # 1-ші класқа (ауру) жату ықтималдығы
```

```python
# МЕТРИКАЛАРДЫ ЕСЕПТЕУ
print("--- 📊 МОДЕЛЬДІҢ НӘТИЖЕЛЕРІ ---")
print(f"Accuracy  (Дәлдік)    : {accuracy_score(y_test, y_pred):.4f} ({(accuracy_score(y_test, y_pred)*100):.1f}%)")
print(f"Precision (Нақтылық)  : {precision_score(y_test, y_pred):.4f}")
print(f"Recall    (Толықтық)  : {recall_score(y_test, y_pred):.4f}")
print(f"F1-Score  (Баланс)    : {f1_score(y_test, y_pred):.4f}\n")
```

```python
# ГРАФИКТЕРДІ САЛУ (Confusion Matrix & ROC-AUC)
fig, ax = plt.subplots(1, 2, figsize=(14, 5))

# А) Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=ax[0],
            xticklabels=['Сау (0)', 'Ауру (1)'],
            yticklabels=['Сау (0)', 'Ауру (1)'])
ax[0].set_title('Confusion Matrix (Шатасу матрицасы)')
ax[0].set_ylabel('Нақты класс (Actual)')
ax[0].set_xlabel('Болжанған класс (Predicted)')

# Ә) ROC-AUC қисығы
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

ax[1].plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.4f})')
ax[1].plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
ax[1].set_xlim([0.0, 1.0])
ax[1].set_ylim([0.0, 1.05])
ax[1].set_xlabel('False Positive Rate (FPR)')
ax[1].set_ylabel('True Positive Rate (Recall)')
ax[1].set_title('ROC-AUC Қисығы')
ax[1].legend(loc="lower right")

plt.tight_layout()
plt.show()
```

### 📝 Модель нәтижелері бойынша қорытынды

Бұл логистикалық регрессия моделі өте жоғары әрі тұрақты нәтиже көрсетті. Барлық негізгі метрикалар **~90-91%** шамасында болуы мынаны білдіреді:

* **Accuracy (90.0%):** Модель жалпы пациенттердің 90%-на дұрыс диагноз қойды. Бұл — өте күшті базалық көрсеткіш (Baseline).
* **Precision (0.9091):** Модель «ауру» деп тапқан адамдардың 91%-ы шынымен де ауру болып шықты. Яғни, «өтірік дабыл» (сау адамды ауру деп қорқыту) өте аз.
* **Recall (0.9091):** Нақты ауру адамдардың 91%-ын модель дұрыс тауып, өткізіп алған жоқ. Медициналық деректер үшін бұл өте маңызды.
* **F1-Score (0.9091):** Precision мен Recall арасында тамаша баланс сақталған. Бұл біздегі деректердің (аурулар мен саулар санының) тепе-тең екенін және модельдің ешқандай ауытқусыз (bias) екі классты да жақсы ажырата алатынын көрсетеді.

**💡 Жалпы түйін:** Модель сәтті оқытылды және жақсы жұмыс істеп тұр. Алайда, бұл медициналық диагноз (жүрек ауруы) болғандықтан, нақты өмірде **Recall** көрсеткішін бұдан да жоғары (мысалы, 95-99%) көтеруге тырысқан жөн. Себебі сау адамды қателесіп емдегеннен гөрі, нақты ауру адамды байқамай қалу (False Negative) анағұрлым қауіптірек.
